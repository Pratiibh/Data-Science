### Problem 3: Exploring data

For each of the questions you proposed in Problem 2, perform an exploratory data analysis designed to address the question. At a minimum, you should produce two visualizations related to each question. Be sure to describe what the visuals show and how they speak to your question of interest.

```python
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
```

```python
sns.violinplot(data=animals, x = "Animal Type", y = "Age", hue = "outcome_is_dead" )
plt.xticks(rotation=45, ha='right')
```

```python
sns.displot(data=animals, x="Animal Type", hue="outcome_is_dead")
plt.xticks(rotation=45, ha='right')
```

The above visualizations show how animal type and age correlate with the overall outcome of the animal.  We can see from the first visualization above many interesting points.  For example, solely from this visualization, I found that almost all Livestock animals above the age of 10 were dead when they were found, but under the age of 10 were all alive when found.  Dogs that were puppies (first 2 years of their life) were extremely less likely to be found dead while dogs above the age of 10 were more likely to be found dead.  Rabbits being found dead or alive looks like a very similar distribution.  The first visualization lacks a quantity being put into our heads however, so the second visualization helps paint the picture a bit better.  The second visualization shows that just under 12000 dogs were found alive and just short of 1000 were found dead.  Because of the second visualization we can see about 15000 cats were lost but found alive, while 4000 were lost and found dead.  Because of this second visualization we can see that cats are the most likely animal to be lost, followed by dogs, and then birds! Livestock was extremely rare to be reported lost, as one would assume as livestock are usually kept gated as they relate directly to the production of a farmers assets, while house animals and pets that aren't tied into economical production were much more likely to be reported lost.

Below are the visualizations to help answer my second question "How does the intake condition of an animal affect its outcome?"

```python
sns.histplot(data=animals, x="Intake Condition", hue="outcome_is_dead")
plt.xticks(rotation=45, ha='right')
```

```python
sns.catplot(data=animals, x = "Animal Type", y = "Age", hue="Intake Condition")
plt.xticks(rotation=45, ha='right')
```

The two visualizations above help to answer my second question "How does the intake condition of an animal affect its outcome?". The first visualization shows the sheer numbers of how many animals are found dead/alive dependent on their intake condition.  For example, we can find that most animals are normal in intake conditions end up surviving. And while many animals that are initially underage/weight survive, many of them die due to this intake condition.  The leading correlation of deaths with intake conditions was that where an animal is found to be underage/weight as an intake condition.  However, looking at per capita results, we can determine that severe illnesses or severe injuries as an intake condition in lost animals meant that the overwhelming majority probability was to find them dead.  The second visualization correlates age more with the intake conditions and shows us a multitude of information about different animals and their intake conditions versus age.  For example, we can see from the second visualization that the majority of rabbits, reptiles, and guinea pigs, are found to be in a normal intake condition, regardless of their age.  Wild animals and birds are most likely to be found with severe injuries, and are most likely to be under 10 years of age.  Dogs and cats are most reliable to be alive between the age of 10-20.  We can see that many dogs are normal on intake condition basis, but dogs near the age of 20 are most likely to be found aged, or injured severely.  However, this pales in comparison to cats, where cats are found to be extremely likely to be severely injured, or severely ill, with a heavy majority of these conditions to be found on cats of ages 3-20.
