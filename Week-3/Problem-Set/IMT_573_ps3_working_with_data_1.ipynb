#### (d) Intake seasonality

One may wonder whether intakes are in any way related to weather. We do not have weather information here but let’s analyze how intakes are related to seasons. Which seasons have the most intakes? Why might this be the case? In your communication of your analysis use one graphical visualization and one tabular respresentation of your findings.

```python
data_filtered = data.copy()
```

```python
data_filtered = data_filtered[data_filtered.Age >= 0]
```

```python
def get_season(date_string):
    month = int(date_string[5:7])
    season = "Autumn"
    if month in [12, 1, 2]:
        season = "Winter"
    elif month in [3, 4, 5]:
        season = "Spring"
    elif month in [6, 7, 8]:
        season = "Summer"
    return season
```

```python
data_filtered['Season'] = data_filtered['Intake Date'].apply(get_season)
```

```python
data_filtered['Season'].value_counts()
```

```python
tabulated = data_filtered.groupby(['Season', 'Animal Type'])['Animal ID'].count().unstack().fillna(0)
tabulated = round(tabulated / tabulated.sum() * 100, 2)
tabulated
```

```python
data_filtered['intake_year_month'] = data_filtered['Intake Date'].str.slice(0,7)
```

```python
data_grouped = data_filtered.groupby(['intake_year_month', 'Season'])['Animal ID'].count().reset_index()
xticks = data_grouped.intake_year_month.unique()
```

```python
plt.figure(figsize=(10,6))

fig = sns.lineplot(data = data_grouped, x = 'intake_year_month',
                  y = 'Animal ID', color = 'grey', alpha = 0.8);
fig = sns.scatterplot(data = data_grouped, x = 'intake_year_month',
                  y = 'Animal ID', hue = 'Season', alpha = 0.8);
plt.xticks(rotation=90);
fig.set_xticks(xticks[1::2]);
plt.ylabel('Number of Animals')
plt.xlabel('Intake Month')
plt.title('Animal Intakes by Month')
sns.move_legend(fig, "upper left", bbox_to_anchor=(1, 1))
```

Summer had much higher intake rates than other season and then a decline across Autumn before a lull in winter.  Interestingly, intake numbers dropped drastically during COVID lockdown months and have been slower to recover since.

#### (e) Decreases during COVID

Did all animal types see a decrease in intakes during 2020 compared to their pre-2020 annual average? Which animal types saw the largest numerical drop and the largest percent drop?

```python
data_prob_3 = data[(data['Jurisdiction'] == 'LONG BEACH') & (data['Outcome Type'] == 'RETURN TO OWNER')]
```

```python
data_prob_3['intake_year'] = data_prob_3['Intake Date'].str.slice(0, 4).astype(int)
```

```python
data_prob_3['intake_year']
```

```python
pre_covid = data_prob_3[data_prob_3.intake_year < 2020].groupby(['Animal Type', 'intake_year'])['Animal ID'].count().reset_index()
pre_covid = pre_covid.groupby(['Animal Type'])['Animal ID'].mean()
pre_covid
```

```python
mid_covid = data_prob_3[data_prob_3.intake_year == 2020].groupby(['Animal Type'])['Animal ID'].count()
mid_covid
```

```python
diffs = pd.DataFrame(pre_covid)

diffs.columns = ['pre_2020']

diffs['2020'] = mid_covid

diffs = diffs.fillna(0)

diffs['diff'] = diffs['pre_2020'] - diffs['2020']

diffs['perc_drop'] = diffs['diff'] / diffs['pre_2020'] * 100

diffs
```
